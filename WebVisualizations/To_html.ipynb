```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
weather_data = "Resources/Weather_Data.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
weather_data_df = pd.read_csv(weather_data, encoding="ISO-8859-1")
```

```python
# Show just the header
weather_data_df.head()
```

```python
# Reorganizing the columns using double brackets
organized_df = weather_data_df[["City","Country","Temperature","Humidity", "Clouds", "Windspeed", "Latitude", "Longitude"]]
organized_df.head()
```

```python
# convert file to html
organized_df.to_html('Weather_Data_PD.html')
```
